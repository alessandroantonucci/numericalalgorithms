```python
def f(x):
  return x*x-3

# Dati a=0 e b=3 interare 5 volte la bisezione
# importing the required modules 
import matplotlib.pyplot as plt 
import numpy as np 

def plotfunction(a,b):
  delta = 0.5
  xmin = 0-delta
  xmax = 3+delta
  x = np.arange(0, 3, 0.1) 
  y = [f(xx) for xx in x]
  # potting the points 
  plt.plot(x, y) 
  plt.xlabel('x')
  plt.ylabel('y')
  plt.title('Ibrido Bisezione - Regula Falsi')
  plt.grid(True)
  plt.arrow(xmin, 0, xmax-0.1, 0, head_width=0.5, head_length=0.1, color='black',lw=4)
  # function to show the plot 
  plt.plot([a], [0], marker='o', markersize=13, color="red")
  plt.plot([b], [0], marker='o', markersize=13, color="blue")
  plt.plot([1.71], [0], marker='o', markersize=13, color="black")
  plt.show()

a = 0
b = 3
if f(a)*f(b)<0:
  for k in range(10):
    plotfunction(a,b)
    c = (a+b)/2
    if f(a)*f(c)<0:
      b = c
    else:
      a = c
```
